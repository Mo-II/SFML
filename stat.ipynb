Load the dataset in a dataframe

```python
import pandas as pd
df = pd.read_csv("ObesityDataSet.csv")
```

Plot a bar chart to show the distribution of obesity levels.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
# Obesity Level (NObeyesdad)
plt.figure(figsize=(8, 6))
sns.countplot(df['NObeyesdad'])
plt.title('Obesity Level')
plt.show()
```

Create a count plot to visualize the distribution of obesity levels by gender.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='NObeyesdad', hue='Gender', palette='coolwarm')
plt.title('Gender Distribution in Each Obesity Level')
plt.xlabel('Obesity Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

Plot a box plot to show the age distribution across different obesity levels.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='NObeyesdad', y='Age', palette='Set2')
plt.title('Age Distribution in Each Obesity Level')
plt.xlabel('Obesity Level')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```

Scatter plot to see the distribution of height and weight colored by obesity level.

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Height', y='Weight', hue='NObeyesdad', palette='viridis', alpha=0.7)
plt.title('Height and Weight Distribution by Obesity Level')
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.legend(title='Obesity Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Visualize the relationship between family history of overweight and obesity levels.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='NObeyesdad', hue='family_history_with_overweight', palette='pastel')
plt.title('Family History of Overweight and Obesity Levels')
plt.xlabel('Obesity Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Family History')
plt.show()
```

```python
# Group by 'CALC' and calculate the mean 'Weight'
CALC_weight_avg = df.groupby('CALC').agg({'Weight': 'mean'}).reset_index()

fig, ax = plt.subplots()
bars = ax.bar(CALC_weight_avg['CALC'], CALC_weight_avg['Weight'])

ax.set_title('Average Weight by Alcohol Consumption')
ax.set_xlabel('Alcohol Consumption')
ax.set_ylabel('Average Weight')
plt.show()
```

```python

# Group by 'CALC' and calculate the mean 'TUE'
calc_tue_avg = df.groupby('CALC').agg({'TUE': 'mean'}).reset_index()

fig, ax = plt.subplots()
bars = ax.bar(calc_tue_avg['CALC'], calc_tue_avg['TUE'])
ax.set_title('Average Time Spent on Activities by Alcohol Consumption')
ax.set_xlabel('Alcohol Consumption')
ax.set_ylabel('Average Time Spent on Activities')
plt.show()
```
